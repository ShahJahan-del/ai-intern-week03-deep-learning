```python
from transformers import AutoTokenizer

# 1. Load the official BERT tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# 2. Tokenize a sample sentence with padding and truncation enabled
text = "Tokenization is fast and easy!"
inputs = tokenizer(text, padding=True, truncation=True, return_tensors="pt")

print("--- BERT TOKENIZATION DEMO ---")
# Show the subword tokens
tokens = tokenizer.tokenize(text)
print(f"1. Subword Tokens: {tokens}")

# Show the actual numerical IDs sent to the model GPU
print(f"2. Input IDs:      {inputs['input_ids'][0].tolist()}")

# Show the Attention Mask (1 for real tokens, 0 for padding)
print(f"3. Attention Mask: {inputs['attention_mask'][0].tolist()}")
```
